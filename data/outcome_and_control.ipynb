```python
outcome_lst = ['onset2COWCS', 'd2incidenceU', 'd3_6incidenceU', 'onsetUCS', 'coup', 'periregular', 'milexp_pergdpSIPRI','milexp_pergdpSIPRI_diff']
```

In order to compare the variation of variables with different scales, you can use the coefficient of variation (CV). The coefficient of variation is a dimensionless measure of relative variability that is calculated by dividing the standard deviation (SD) of a variable by its mean (µ):

CV = (SD / µ) * 100

By calculating the CV for each variable, you can compare their relative variation and choose the one with the highest CV as the variable with the most variation.

```python
def compute_cv(variable):
    # Convert the input list to a numpy array
    variable_array = np.array(variable)

    # Remove NaN values
    variable_array = variable_array[~np.isnan(variable_array)]

    # Calculate mean, standard deviation, and CV
    mean = np.mean(variable_array)
    std_dev = np.std(variable_array)
    cv = (std_dev / mean) * 100

    return cv
```

```python
for outcome in outcome_lst:
    print(outcome, str(compute_cv(df[outcome])))
```

```python
# Plot histograms for each column
for column in outcome_lst:
    arr = df[column]
    arr = arr[~np.isnan(arr)]
    plt.figure()
    plt.hist(arr, bins=10, edgecolor='black', alpha=0.7)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()
```
