# Main results for the paper

```python
import pandas as pd
```

```python
mix_res = pd.read_csv('mix_res.csv')
forward_res = pd.read_csv('forward_res.csv')
backward_res = pd.read_csv('backward_res.csv')
```

```python
stacked_df = pd.concat([mix_res, forward_res, backward_res], axis=0)
final = stacked_df.reset_index(drop=True)
final
```

```python
paper_res = final[
    (final['treatment'] == 'treat_fuel') &
    (final['instrument'] == 'iv_transport') &
    ~(final['fixed_effects'] & final['stationary_controls'])
]

# Print the selected rows
paper_res = paper_res.reset_index(drop=True)
```

```python
paper_res = paper_res.drop(['outcome', 'treatment', 'instrument'], axis=1)
paper_res
```

```python
desired_columns = ['gvc_type', 'fixed_effects', 'stationary_controls', 'tau_hat', 'std_hat', 'p_val']
paper_res = paper_res[desired_columns]
paper_res
```
